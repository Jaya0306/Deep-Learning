```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
# Define the error function
def compute_error(X, Y, w, b):
    err = 0.0
    for x, y in zip(X, Y):
        fx = sigmoid(w * x + b)
        err += 0.5 * (fx - y) ** 2  # Squared error
    return err
```

```python
# Compute gradients for w and b
def grad_w(w, b, x, y):
    fx = sigmoid(w * x + b)
    return (fx - y) * fx * (1 - fx) * x
```

```python
def grad_b(w, b, x, y):
    fx = sigmoid(w * x + b)
    return (fx - y) * fx * (1 - fx)
```

```python
# Batch Gradient Descent
def gradient_descent(X, Y, eta=1.0, max_epochs=1000):
    w, b = 2, 2  # Initial values
    w_hist, b_hist = [w], [b]
    
    for i in range(max_epochs):
        dw, db = 0, 0
        for x, y in zip(X, Y):
            dw += grad_w(w, b, x, y)
            db += grad_b(w, b, x, y)
        w -= eta * dw / len(X)  # Averaging over batch
        b -= eta * db / len(X)
        w_hist.append(w)
        b_hist.append(b)
    return w_hist, b_hist
```

```python
# Generate data points
X = np.array([0.5, 2.5])
Y = np.array([0.2, 0.9])
```

```python
# Create contour plot for loss function
def create_plot(w_hist, b_hist, title):
    w_vals = np.linspace(-4, 4, 100)
    b_vals = np.linspace(-4, 4, 100)
    W, B = np.meshgrid(w_vals, b_vals)
    Z = np.array([[compute_error(X, Y, w, b) for w, b in zip(W_row, B_row)] for W_row, B_row in zip(W, B)])
    plt.figure(figsize=(10, 6))
    plt.contourf(W, B, Z, levels=50, cmap="coolwarm")
    plt.colorbar(label="Loss")
    # Plot gradient descent paths
    plt.plot(w_hist, b_hist, color="black")
    plt.xlabel("w")
    plt.ylabel("b")
    plt.title(title)
    plt.show()
```

```python
# Compute gradient paths
w_hist_gd, b_hist_gd = gradient_descent(X, Y, eta=1.0, max_epochs=50)
create_plot(w_hist_gd, b_hist_gd,"Gradient Descent")
```

```python
# Stochastic Gradient Descent
def stochastic_gradient_descent(X, Y, eta=1.0, max_epochs=1000):
    w, b = 2, 2  # Initial values
    w_hist, b_hist = [w], [b]

    for i in range(max_epochs):
        for x, y in zip(X, Y):
            dw = grad_w(w, b, x, y)
            db = grad_b(w, b, x, y)
            w -= eta * dw  # Update immediately per sample
            b -= eta * db
            w_hist.append(w)
            b_hist.append(b)
    return w_hist, b_hist
```

```python
w_hist_sgd, b_hist_sgd = stochastic_gradient_descent(X, Y, eta=1.0, max_epochs=50)
create_plot(w_hist_sgd, b_hist_sgd,"Stochastic Gradient Descent")
```

```python
# Mini-Batch Gradient Descent
def mini_batch_gradient_descent(X, Y, eta=1.0, max_epochs=1000, batch_size = 2):
    w, b = 2, 2  # Initial values
    w_hist, b_hist = [w], [b]
    n = len(X)

    for _ in range(max_epochs):
        indices = np.random.permutation(n)  # Shuffle data
        X_shuffled, Y_shuffled = X[indices], Y[indices]

        for i in range(0, n, batch_size):
            X_batch, Y_batch = X_shuffled[i:i + batch_size], Y_shuffled[i:i + batch_size]
            dw = np.mean([grad_w(w, b, x, y) for x, y in zip(X_batch, Y_batch)])
            db = np.mean([grad_b(w, b, x, y) for x, y in zip(X_batch, Y_batch)])
            w -= eta * dw
            b -= eta * db
            w_hist.append(w)
            b_hist.append(b)

        dw, db = 0,0

    return w_hist, b_hist
```

```python
w_hist_mgd, b_hist_mgd = mini_batch_gradient_descent(X, Y, eta=1.0, max_epochs=50)
create_plot(w_hist_mgd, b_hist_mgd,"Mini Batch Gradient Descent")
```
