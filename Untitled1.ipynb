```python
import torch
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the Rosenbrock function: a non-convex loss function
def rosenbrock(x, y):
    return (1 - x) ** 2 + 100 * (y - x**2) ** 2
```

```python
# Compute the loss surface
x_vals = np.linspace(-2, 2, 100)
y_vals = np.linspace(-1, 3, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = rosenbrock(X, Y)
```

```python
# Define optimizers
optimizers = {
    "SGD": lambda params: optim.SGD(params, lr=0.001),
    "Momentum": lambda params: optim.SGD(params, lr=0.001, momentum=0.9),
    "NAG": lambda params: optim.SGD(params, lr=0.001, momentum=0.9, nesterov=True),
    "Adam": lambda params: optim.Adam(params, lr=0.01),
    "RMSprop": lambda params: optim.RMSprop(params, lr=0.01)
}
```

```python
# Initialize in a HIGH ERROR region (top-left of the plot)
init_x, init_y = -3.5, 3.5  # This is inside a red high-error zone
```

```python
# Track paths for different optimizers
paths = {}
```

```python
for name, opt_func in optimizers.items():
    x = torch.tensor([init_x], requires_grad=True)  # Start far from the minimum
    y = torch.tensor([init_y], requires_grad=True)
    optimizer = opt_func([x, y])
    
    path = []
    for _ in range(100):  # Iterate for 100 steps
        optimizer.zero_grad()
        loss = rosenbrock(x, y)
        loss.backward()
        optimizer.step()
        path.append((x.item(), y.item()))
    
    paths[name] = np.array(path)
```

```python
# Plot contour and optimizer paths
plt.figure(figsize=(10, 6))
plt.contour(X, Y, Z, levels=np.logspace(-1, 3, 20), cmap="coolwarm", alpha=0.7)  # Complex contour plot

# Plot optimization paths
for name, path in paths.items():
    plt.plot(path[:, 0], path[:, 1], marker="o", markersize=3, label=name)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Contour Plot of Rosenbrock Function with Optimizer Paths")
plt.legend()
plt.show()
```
