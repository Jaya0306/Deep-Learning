An activation function is a mathematical function applied to the weighted sum of inputs in a neural network neuron. It determines whether a neuron should be activated (fire a signal) or not. Without activation functions, neural networks would behave like linear regression models, unable to capture complex patterns.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Step function

The step function, also known as the threshold function, is the simplest activation function in machine learning. It determines whether a neuron should be activated based on a predefined threshold. If the input is greater than or equal to the threshold, the neuron fires (outputs 1). Otherwise, it remains inactive (outputs 0).

```python
def step_function(x):
    return np.where(x >= 0, 1, 0)
```

```python
# Generate input values
x = np.linspace(-10, 10, 1000)
y = step_function(x)
```

```python
# Plot the step function
plt.figure(figsize=(6, 4))
plt.plot(x, y, label="Step Function", color='blue')
plt.axhline(y=0, color='gray', linestyle='--', linewidth=0.5)
plt.axhline(y=1, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(x=0, color='red', linestyle='--', linewidth=0.5, label="Threshold (x=0)")
plt.xlabel("Input (x)")
plt.ylabel("Output (f(x))")
plt.title("Step Function")
plt.legend()
plt.grid()
plt.show()
```

## Sigmoid function

The Sigmoid function is one of the most fundamental activation functions in neural networks. It is commonly used in binary classification problems and is known for converting any real-valued number into a value between 0 and 1, making it particularly useful for probabilistic outputs.

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# Generate values for plotting
x = np.linspace(-10, 10, 1000)
y = sigmoid(x)
```

```python
# Plot sigmoid function
plt.figure(figsize=(6,4))
plt.plot(x, y, label="Sigmoid Function", color='blue')
plt.axhline(y=0, color='gray', linestyle='--', linewidth=0.5)
plt.axhline(y=1, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(x=0, color='red', linestyle='--', linewidth=0.5, label="x=0 (midpoint)")
plt.xlabel("Input (x)")
plt.ylabel("Output (f(x))")
plt.title("Sigmoid Activation Function")
plt.legend()
plt.grid()
plt.show()
```

## Tanh function

The Tanh (Hyperbolic Tangent) function is a widely used activation function in neural networks, especially in hidden layers. It is similar to the Sigmoid function, but it is 
zero-centered, which improves training efficiency.

```python
def tanh(x):
    return np.tanh(x)
```

```python
# Generate values for plotting
x = np.linspace(-10, 10, 1000)
y = tanh(x)
```

```python
# Plot tanh function
plt.figure(figsize=(6,4))
plt.plot(x, y, label="Tanh Function", color='blue')
plt.axhline(y=0, color='gray', linestyle='--', linewidth=0.5)
plt.axhline(y=1, color='gray', linestyle='--', linewidth=0.5)
plt.axhline(y=-1, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(x=0, color='red', linestyle='--', linewidth=0.5, label="x=0 (midpoint)")
plt.xlabel("Input (x)")
plt.ylabel("Output (f(x))")
plt.title("Tanh Activation Function")
plt.legend()
plt.grid()
plt.show()
```

## ReLU function

The Rectified Linear Unit (ReLU) function is the most commonly used activation function in deep learning, particularly in Convolutional Neural Networks (CNNs) and deep neural networks (DNNs). It overcomes the limitations of Sigmoid and Tanh by avoiding vanishing gradients and enabling faster training.

```python
def relu(x):
    return np.maximum(0, x)
```

```python
# Generate values for plotting
x = np.linspace(-10, 10, 1000)
y = relu(x)
```

```python
# Plot ReLU function
plt.figure(figsize=(6,4))
plt.plot(x, y, label="ReLU Function", color='blue')
plt.axhline(y=0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(x=0, color='red', linestyle='--', linewidth=0.5, label="x=0 (threshold)")
plt.xlabel("Input (x)")
plt.ylabel("Output (f(x))")
plt.title("ReLU Activation Function")
plt.legend()
plt.grid()
plt.show()
```

## Leaky ReLU function

The Leaky Rectified Linear Unit (Leaky ReLU) is an improved version of ReLU that addresses its main limitation—the Dying ReLU problem. It allows small negative values instead of zero for negative inputs, preventing neurons from completely deactivating.

```python
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)
```

```python
# Generate values for plotting
x = np.linspace(-10, 10, 1000)
y = leaky_relu(x)
```

```python
# Plot Leaky ReLU function
plt.figure(figsize=(6,4))
plt.plot(x, y, label="Leaky ReLU (α=0.01)", color='blue')
plt.axhline(y=0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(x=0, color='red', linestyle='--', linewidth=0.5, label="x=0 (threshold)")
plt.xlabel("Input (x)")
plt.ylabel("Output (f(x))")
plt.title("Leaky ReLU Activation Function")
plt.legend()
plt.grid()
plt.show()
```

## PReLU function

Parametric ReLU (PReLU) is a modification of the Leaky ReLU activation function where the negative slope (α) is learned during training rather than being a fixed value.

```python
def prelu(x, alpha):
    return np.where(x > 0, x, alpha * x)
```

```python
x = np.linspace(-5, 5, 100)
y_prelu = prelu(x, alpha=0.2)  # PReLU (learnable alpha)
```

```python
# Plot Parametric ReLU function
plt.figure(figsize=(6,4))
plt.plot(x, y_prelu, label="PReLU (α=0.2, learned)", linestyle='-', color='green')
plt.axhline(y=0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(x=0, color='black', linestyle='--', linewidth=0.5)
plt.xlabel("Input (x)")
plt.ylabel("Output (f(x))")
plt.title("Parametric ReLU Activation Function")
plt.legend()
plt.grid()
plt.show()
```

## RReLU function

Randomized ReLU (RReLU) is a variant of the Leaky ReLU activation function, where the negative slope (α) is randomly chosen from a given range during training but remains fixed during inference.

```python
def rrelu(x, alpha_min, alpha_max):
    alpha = np.random.uniform(alpha_min, alpha_max, size=x.shape)
    return np.where(x > 0, x, alpha * x)
```

```python
x = np.linspace(-5, 5, 100)
y_rrelu = rrelu(x, 0.1, 0.3)  # RReLU (random alpha)
```

```python
# Plot Randomized ReLU function
plt.figure(figsize=(6,4))
plt.plot(x, y_rrelu, label="RReLU (α random 0.1-0.3)", linestyle='-', color='purple')
plt.axhline(y=0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(x=0, color='black', linestyle='--', linewidth=0.5)
plt.xlabel("Input (x)")
plt.ylabel("Output (f(x))")
plt.title("# Randomized ReLU Activation function")
plt.legend()
plt.grid()
plt.show()
```

## ELU function

The Exponential Linear Unit (ELU) is an activation function that helps solve the vanishing gradient problem while improving learning speed.

```python
def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))
```

```python
x = np.linspace(-3, 3, 100)
y = elu(x, alpha=1.0)
```

```python
# Plot ELU function
plt.figure(figsize=(6,4))
plt.plot(x, y, label="ELU (α=1.0)", color='blue')
plt.axhline(0, color='black', linewidth=0.5, linestyle="--")
plt.axvline(0, color='black', linewidth=0.5, linestyle="--")
plt.title("Exponential Linear Unit (ELU) Activation Function")
plt.xlabel("Input (x)")
plt.ylabel("Output (f(x))")
plt.legend()
plt.grid()
plt.show()
```

## Softmax function

The Softmax activation function is widely used in multi-class classification problems. It converts raw logits (unbounded scores) into probabilities that sum up to 1, making it ideal for categorical outputs.

```python
def softmax(x):
    x_exp = np.exp(x - np.max(x))  # Stability trick
    return x_exp / np.sum(x_exp)
```

```python
# Example
logits = np.array([2.0, 1.0, 0.1])
print(softmax(logits))  # Output: [0.659, 0.242, 0.099]
```

## Comparison of Activation Functions

```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def step(x):
    return np.where(x >= 0, 1, 0)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def prelu(x, alpha=0.1):  # Parameterized ReLU
    return np.where(x > 0, x, alpha * x)

def rrelu(x, l=0.1, u=0.3):
    alpha = np.random.uniform(l, u, size=x.shape)
    return np.where(x > 0, x, alpha * x)

def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def softmax(x):
    exp_x = np.exp(x - np.max(x))  # Subtract max for numerical stability
    return exp_x / np.sum(exp_x)

# Generate input values
x = np.linspace(-5, 5, 400)

# Apply activation functions
activations = {
    "Step": step(x),
    "Sigmoid": sigmoid(x),
    "Tanh": tanh(x),
    "ReLU": relu(x),
    "Leaky ReLU": leaky_relu(x),
    "PReLU": prelu(x),
    "RReLU": rrelu(x),
    "ELU": elu(x),
}

# Plot activation functions
plt.figure(figsize=(8, 4))
for name, y in activations.items():
    plt.plot(x, y, label=name)

plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Comparison of Activation Functions")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.legend()
plt.grid(True)
plt.show()
```
